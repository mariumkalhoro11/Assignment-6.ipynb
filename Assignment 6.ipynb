# ===== QUESTION 4 =====
# Print n^2 for n = 1 to 5

print("\nQ4:")
for n in range(1, 6):
    print(n**2)

# Output:
# 1, 4, 9, 16, 25


# Alternative (list comprehension)
print("\nQ4 Alternative:")
[print(n**2) for n in range(1, 6)]


# ===== QUESTION 5 =====
# Loop until i^2 >= 1000

print("\nQ5:")

i = 1

while i**2 < 1000:
    i += 1

print(i)  # 32

# Explanation:
# 31^2 = 961 (less than 1000)
# 32^2 = 1024 (greater than 1000)
# So final i = 32
